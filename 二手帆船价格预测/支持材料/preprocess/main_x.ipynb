```python
# Basic Analytical Package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Read original df
data=pd.read_excel('Monohulled_coastline_avg_Numerical.xlsx',index_col=0)
data
```

```python
# 目标变量分布
sns.displot(np.log(data['Listing Price (USD)']),kde=True, rug=True)
plt.show()

print("偏度为 e^%f " % np.log(data['Listing Price (USD)'].skew()))
print("峰度为 e^%f " % np.log(data['Listing Price (USD)'].kurt()))
```
